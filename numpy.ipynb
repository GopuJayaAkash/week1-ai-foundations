```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print(a)
print(b)
```

```python
print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Element-wise multiplication
print(a / b)   # Element-wise division
print(a ** 2)  # Power
```

```python
print(a + 10)  #Scalar addition
print(a * 2)   #Scalar multiplication
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print(arr.shape)   # Dimensions
print(arr.ndim)    # Number of dimensions
print(arr.size)    # Total elements
print(arr.dtype)   # Data type
```

```python
print(a[0])      # First element
print(a[-1])     # Last element
print(a[1:3])    # Slice
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])

reshaped = arr.reshape(2, 3)
print(reshaped)
```

```python
print(np.sum(a))
print(np.mean(a))
print(np.max(a))
print(np.min(a))
print(np.std(a))
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print(A @ B)          # Matrix multiplication
print(np.transpose(A))
```

```python
np.zeros((2, 3))   # All zeros
np.ones((2, 3))    # All ones
np.eye(3)          # Identity matrix
np.arange(0, 10, 2)
np.linspace(0, 1, 5)
```

```python
print(a > 2)
print(a[a > 2])   # Filtering
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print(arr + 10)
```
